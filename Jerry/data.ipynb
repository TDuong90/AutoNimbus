```python
import pandas as pd
```

```python
frame = pd.read_csv('./Data_from_2010_to_present.csv')
frame.head()
```

```python
frame.dtypes
```

```python
frame['Date Occurred'] = pd.to_datetime(frame['Date Occurred'], infer_datetime_format = True)
frame.dtypes
```

```python
print(frame.count())
frame.dropna(inplace=True)
print(frame.count())
```

```python
grouped_day_frame = frame.groupby("Date Occurred").count()
grouped_day_frame.head(2)
```

```python
day_frame = grouped_day_frame['Location']
print(type(day_frame))
print(day_frame)
# day_frame.to_csv('./count_per_day.csv')
```

```python
frame_day = pd.DataFrame({'count':day_frame, 'year':'', 'month':''})
frame_day
```

```python
for index, row in frame_day.iterrows():
    print(index.strftime('%Y'), index.strftime('%m'));
    frame_day.loc[index,'year'] = index.strftime('%Y');
    frame_day.loc[index,'month'] = index.strftime('%m');
```

```python
frame_day
```

```python
frame_per_month = frame_day.groupby(['year', 'month']).sum()
frame_per_month.head()
```

```python
# frame_per_month.to_csv('count_per_month.csv')
```

```python
frame_per_month['date']=''
frame_per_month
```

```python
for index, row in frame_per_month.iterrows():
    print(index[0])
    print(index[1])
    print(row)
    print('*********************************')
    frame_per_month.loc[index,'date'] = f'{index[0]}-{index[1]}-01';
```

```python
frame_per_month
```

```python
count_per_month = frame_per_month[['date', 'count']]
pd.to_datetime(count_per_month['date'], infer_datetime_format = True)
```

```python
count_per_month.to_csv('count_per_month.csv')
```
